**1a)** The rottentomatoes web page **THE BEST MOVIES OF 2021, RANKED BY TOMATOMETER** (https://editorial.rottentomatoes.com/guide/2021-best-movies/) lists the top 100 movies of this year.  To avoid issues accessing the website, the contents of this page have been saved to a file *rotten.txt* and the code to retrieve the contents of this file is provided below. 

Use these contents to scrape the rank, name, tmeterscore, director and starring for each of the movies and populate a dataframe **movies** with columns *Rank, Name, tMeterScore, Director, Cast*.  You must return the dataframe sorted by *Rank* (from 1 down to the lowest) **[25 Points]**

***Useful Hints***

To find a tag named 'foo' with attribute 'class' having a value that starts with 'Bar' use:<br>
find ('foo',{'class': re.compile('Bar.*')})

To extract the first number from a string *mystring* use:<br>
re.findall ('\d\*', *mystring*)[0]






```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup


#read the contents of the file rotten.txt 
with open('rotten.txt', 'rb') as f: 
    content = f.read()
results_page = BeautifulSoup(content,'lxml') 
name_list = results_page.find_all('div',{'class':"article_movie_title"})
name = list()
for i in name_list:
    name.append(i.find('a').get_text())
    
```

```python
rank_list = results_page.find_all('div',{'class':"countdown-index"})
rank = list()
for i in rank_list:
    rank.append(i.get_text())

for j in range(0,len(rank)):
    rank[j]=rank[j][1:]
    
rank
```

```python
tmeter_list = results_page.find_all('span',{'class':"tMeterScore"})
tmeter = list()
for i in tmeter_list:
    tmeter.append(i.get_text())

tmeter
```

```python
director_list = results_page.find_all('div',{'class':"info director"})
director = list()
for i in director_list:
    try:
        director.append(i.find('a').get_text())
    except:
        director.append('')
director
#director_list[0].find('a').get_text()
```

```python
cast_list = results_page.find_all('div',{'class':"info cast"})
cast = list()
for i in cast_list:
    try:
        cast.append(i.find('a').get_text())
    except:
        cast.append('')
cast
```

```python
import pandas as pd
import numpy as np

attributes = [rank,name,tmeter,director,cast]
df = pd.DataFrame(attributes).transpose()
df.columns=['Rank','Name','Tmeterscore','Director','Cast']
df.sort_values(by = 'Rank')
```

```python
df.sort_values(by = 'Rank')
```

```python
import matplotlib.pyplot as plt
tmt=list()
for h in range(0,len(df['Tmeterscore'])):
    tmt.append(df['Tmeterscore'][h][:-1])
tmt

plt.plot(tmt)
```
